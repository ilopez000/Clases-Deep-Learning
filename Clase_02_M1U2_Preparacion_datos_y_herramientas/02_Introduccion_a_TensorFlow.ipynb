## 3. Crear tensores: constantes y variables

Toda la información en TF se almacena en **tensores**. Hay dos tipos:

- **Constantes** (`tf.constant`): inmutables.
- **Variables** (`tf.Variable`): mutables; su valor cambia durante el entrenamiento (p. ej. los pesos de una red).

```python
C = tf.constant(0)
print("Constante:", C.numpy())

VA = tf.Variable([[1.0, 2.0],
                  [3.0, 4.0]])
print("Variable VA (shape", VA.shape, "):")
print(VA.numpy())

# Una variable puede modificarse; una constante no:
VA.assign(VA * 2)
print("VA tras assign (x2):")
print(VA.numpy())
```

## 4. Generar tensores: zeros, ones y random

Crear tensores grandes a mano es tedioso. Estas funciones lo simplifican.
`random` admite una **semilla** para resultados reproducibles.

```python
ceros = tf.zeros([2, 3])
unos = tf.ones([2, 3])

tf.random.set_seed(42)
aleatorio = tf.random.uniform([2, 3], minval=0, maxval=1)

print("zeros([2,3]):\n", ceros.numpy())
print("\nones([2,3]):\n", unos.numpy())
print("\nrandom.uniform([2,3]):\n", aleatorio.numpy().round(3))
```

## 5. Operaciones sobre tensores

- **`concat`**: combina tensores (de tamaños compatibles).
- **`reshape`**: cambia la forma (el nº total de elementos debe coincidir).
- **`cast`**: cambia el tipo de dato (p. ej. de decimal a entero).

```python
a = tf.constant([[1, 2], [3, 4]])
b = tf.constant([[5, 6], [7, 8]])

print("concat (axis=0):\n", tf.concat([a, b], axis=0).numpy())

cuadrada = tf.constant([[1, 2], [3, 4]])
columna = tf.reshape(cuadrada, [1, 4])
print("\nreshape 2x2 -> 1x4:", columna.numpy())

decimales = tf.constant([1.7, 2.3, 3.9])
enteros = tf.cast(decimales, tf.int32)
print("\ncast float -> int:", enteros.numpy())
```

## 6. Calcular gradientes con GradientTape

Entrenar una red consiste, en esencia, en calcular **gradientes** para ajustar los coeficientes.
`tf.GradientTape` registra las operaciones y permite derivarlas automáticamente.

Lo probamos con la **tangente hiperbólica** (`tanh`), una función de activación habitual.
Su derivada es `1 - tanh(x)²`.

```python
x = tf.Variable(0.5)

with tf.GradientTape() as tape:
    y = tf.math.tanh(x)

dy_dx = tape.gradient(y, x)

print("x          =", x.numpy())
print("tanh(x)    =", y.numpy().round(5))
print("derivada   =", dy_dx.numpy().round(5))
print("comprobación 1 - tanh(x)^2 =", (1 - y.numpy()**2).round(5))
```

## 7. Operaciones matriciales

La mayoría de los cálculos de una red de neuronas son **operaciones matriciales**. Las más comunes en TF:

```python
A = tf.constant([[1.0, 2.0], [3.0, 4.0]])
B = tf.constant([[5.0, 6.0], [7.0, 8.0]])

print("add (A + B):\n", tf.add(A, B).numpy())
print("\ntranspose(A):\n", tf.transpose(A).numpy())
print("\nmatmul (A x B):\n", tf.matmul(A, B).numpy())
```

```python
print("det(A):", tf.linalg.det(A).numpy().round(3))            # determinante
print("tensordot (producto escalar):",
      tf.tensordot(A, B, axes=2).numpy())                        # escalar
print("\neye(3) - matriz identidad:\n", tf.eye(3).numpy())
```

## Resumen

- En TF todo son **tensores**: constantes (inmutables) y variables (mutables, como los pesos).
- Generamos tensores con `zeros`, `ones` y `random`, y los manipulamos con `concat`, `reshape` y `cast`.
- `GradientTape` calcula **gradientes** automáticamente: la base del entrenamiento de redes de neuronas.
- TF ofrece todas las **operaciones matriciales** (`matmul`, `transpose`, `det`, …) que necesitan las redes.
